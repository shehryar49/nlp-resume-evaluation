```python
def c(text):
    a = text.count("c/c++")
    return len(re.findall("(\\t|\\n|,|\\.| |^)c( |$|\\t|\\n|,|\\.)",text)) + a
def cpp(text):
    a = len(re.findall("(\\t|\\n|,|\\.| |^)cpp( |$|\\t|\\n|,|\\.)",text))
    b = len(re.findall("(\\t|\\n|,|\\.| |^)c\\+\\+( |$|\\t|\\n|,|\\.)",text))
    c = text.count("c/c++")
    return a+b+c
def assembly(text):
    return text.count("assembly")
def java(text):
    return len(re.findall("(\\t||,|\\.\\n| |^)java( |$|\\t|,|\\.|\\n)",text))
def js(text):
    return len(re.findall("(\\t|,|\\.|\\n| |^)js( |$|\\t|,|\\.|\\n)",text))
def objC(text):
    a = text.count("objective c")
    b = text.count("objective-c")
    return a+b
def datastructs(text):
    a = text.count("datastructures")
    b = text.count("data-structures")
    c = text.count("data structures")
    return a+b+c
def socket(text):
    return len(re.findall("(\t|,|\\.|\\n| |^)socket( |$|\\t|\\n|,|\\.)",text))
def sql(text):
    return len(re.findall("(\\t|\\n,|\\.|| |^)sql( |$|\\t|,|\\.|\\n)",text))
def go(text):
    return len(re.findall("(\\t|\\n|,|\\.| |^)go( |$|\\t|\\n|,|\\.)",text))

dims = {
    "c" : 0,
    "cpp": 1,
    "assembly": 2,
    "python": 3,
    "java" : 4,
    "javascript" : 5, 
    "html" : 6,
    "css" : 7,
    "lua" : 8,
    "ruby" : 9,
    "rust" : 10,
    "go" : 11,
    "swift" : 12,
    "objective-c" : 13, 
    "android development" : 14, 
    "ios development" : 15,
    "mobile development" : 16,
    "web development" : 17,
    "machine learning" : 18,
    "data science" : 19,
    "embedded development" : 20, 
    "data structures" : 21,
    "tensorflow" : 22,
    "pytorch" : 23,
    "pandas" : 24,
    "reactjs" : 25,
    "nextjs" : 26,
    "vue" : 27,
    "angular" : 28,
    "bootstrap" : 29,
    "sdl" : 30,
    "flex" : 31,
    "bison" : 32,
    "llvm" : 33,
    "compiler" : 34, 
    "libcurl" : 35,
    "websockets" : 36,
    "socket" : 37,
    "springboot" : 38, 
    "laravel" : 39,
    "django" : 40,
    "fastapi" : 41,
    "flask" : 42,
    "sql" : 43,
    "mongodb" : 44, 
    "mysql" : 45,
    "postgresql" : 46, 
    "mssql" : 47,
    "qt" : 48,
    "php" : 49
}
def generate_embedding(text):
    return [
        c(text),
        cpp(text),
        text.count("assembly"),
        text.count("python"),
        java(text),
        text.count("javascript") + js(text),
        text.count("html"),
        text.count("css"),
        text.count("lua"),
        text.count("ruby"),
        text.count("rust"),
        go(text),
        text.count("swift"),
        objC(text),
        text.count("android"),
        text.count("ios"),
        text.count("mobile development"),
        text.count("web"),
        text.count("machine learning"),
        text.count("data science"),
        text.count("embedded development"),
        datastructs(text),
        text.count("tensorflow"),
        text.count("pytorch"),
        text.count("pandas"),
        text.count("reactjs"),
        text.count("nextjs"),
        text.count("vue"),
        text.count("angular"),
        text.count("bootstrap"),
        text.count("sdl"),
        text.count("flex"),
        text.count("bison"),
        text.count("llvm"),
        text.count("compiler") ,
        text.count("libcurl"),
        text.count("websocket"),
        socket(text),
        text.count("springboot"),
        text.count("laravel"),
        text.count("django"),
        text.count("fastapi"),
        text.count("flask"),
        sql(text),
        text.count("mongodb"),
        text.count("mysql"),
        text.count("postgresql"),
        text.count("mssql"),
        text.count("qt"),
        text.count("php")
    ]
def add_weighted_sum(vec,to_update,keys,weights):
    i = 0
    for key in keys:
        vec[dims[to_update]] += vec[dims[key]] * weights[i]
        i += 1
def add_dep_score(vec):
    # sdl, flex and bison are c libraries
    add_weighted_sum(vec,"c",["sdl","flex","bison"],[0.1] * 3)
    # c does help with cpp
    vec[dims["cpp"]] += 0.4 * vec[dims["c"]]
    # tensorflow, pytorch etc improve machine learning skills
    add_weighted_sum(vec,"machine learning",["pytorch","tensorflow"],[0.5] * 2)
    # bootstrap is css
    add_weighted_sum(vec,"css",["bootstrap"],[0.1])
    # web frameworks like laravel, django, fastapi, flask, reactjs, vue, angular, bootstrap, html, css, javascript improve web dev
    # skills
    add_weighted_sum(vec,"web development",["django","fastapi","laravel","flask","reactjs","vue","angular","bootstrap","html","css","javascript"],[0.2] * 11)
    # python library improve your python skills
    add_weighted_sum(vec,"python",["django","flask","pytorch","tensorflow","pandas"],[0.1] * 5)
    # js libraries imrpove js skills
    add_weighted_sum(vec,"javascript",["reactjs","vue","angular","nextjs"],[0.1] * 4)
    # android/ios development is mobile development
    add_weighted_sum(vec,"mobile development",["android development","ios development"],[0.5]*2)
    # flex,bison and llvm improve your compiler construction skills
    add_weighted_sum(vec,"compiler",["llvm","flex","bison"],[0.7] * 3)
    # qt improves your cpp skills( a little bit)
    add_weighted_sum(vec,"cpp",["qt"],[0.3] * 1)
    # springboot should improve java?
    add_weighted_sum(vec,"java",["springboot"],[0.1])
    # mysql, mssql, postgresql all are sql
    add_weighted_sum(vec,"sql",["mysql","mssql","postgresql"],[0.2]*3)
    #done
    
```

```python
def evaluate_resumes(resume_paths,jd):
    jd = jd.lower()
    jd_vec = generate_embedding(jd)
    for i in range(len(jd_vec)):
        if jd_vec[i]:
            jd_vec[i] = 1
    add_dep_score(jd_vec)
    for path in resume_paths:
        text = ""
        reader = PdfReader(path)
        for page in reader.pages:
            text = text + '\n'+ page.extract_text()
        text = text.lower()
        vec = generate_embedding(text)
        add_dep_score(vec)
        # find cosine similarity
        norm1 = norm(jd_vec)
        norm2 = norm(vec)
        cosine = 0
        if norm1 and norm2:
            cosine = np.dot(jd_vec,vec)/(norm(jd_vec)*norm(vec))
        print(path,cosine)
```
